```python
#Stock Prize Analyzer
print("Goal:   Analyze Stock Price Data and Find returns and moving average")
print("======================================================================")
print("Problem Statement:")
print("-"*70)
print("Given closing prices of a stock, Find:")
print("\t\t1.Daily return Percentage")
print("\t\t2.3-day Moving Average")
print("\t\t3.Plot booth closing Price and Moving Average")
print("-"*70)
import pandas as pd
import matplotlib.pyplot as plt
prices=pd.Series([150,152,149,153,155,157,160],index=pd.date_range("2025-11-10",periods=7))
print("Closing Prices:\n",prices)
#1.Daily return percentage change
returns=prices.pct_change()*100
print("\nDaily Return(%):\n",returns.round(2))
#2.3-day moving average
ma=prices.rolling(window=3).mean()
print("\n3-day Moving Average:\n",ma)
print()
#3.Plot booth closing Price and Moving Average
plt.figure(figsize=(8,5))
plt.plot(prices, label='Closing Price', marker='o')
plt.plot(ma, label='3-Day MA', linestyle='--')
plt.title("Stock Price and Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
